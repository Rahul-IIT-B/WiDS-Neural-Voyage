## Basics of Pandas and Matplotlib


Pandas is a utility which is very helpful in dealing with data in a organized manner, while matplotlib is majorly used for plotting different types of graphs<br><br>
Again, in this assignment, you are allowed to search for syntax related stuff!


```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading the dataset from a csv file


```python
df = pd.read_csv('iris_dataset.csv')
print(df.to_string())
```

Perform the following operations on the dataset:

- Print the dataset with the _sepal length_ values sorted in descending order


```python
df_sorted = df.sort_values('sepal length (cm)',ascending=False)
print(df_sorted.to_string())
```

- Print only the _petal length_ and _petal width_ fields for all data points


```python
print(df[['petal length (cm)', 'petal width (cm)']].to_string())
```

- Extract the first 20 entries of the original dataset and print it


```python
print(df.head(20))
```

- Print only those rows where _sepal length_ is greater than 6


```python
print(df[df['sepal length (cm)'] > 6])
```

- Create line plots for each feature on a single plot, with proper labels and a legend at top-right corner. Label the X-axis as 'Index' and the Y-axis as 'Length (cm)'


```python
plt.figure(figsize=(10, 6))
plt.plot(df['sepal length (cm)'], label='Sepal Length (cm)')
plt.plot(df['sepal width (cm)'], label='Sepal Width (cm)')
plt.plot(df['petal length (cm)'], label='Petal Length (cm)')
plt.plot(df['petal width (cm)'], label='Petal Width (cm)')
plt.plot(df['target'], label='Target')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.title('Line Plots of Features')
plt.legend(loc='upper right')
plt.show()
```

Now create a scatter plot between _petal length_ and _petal width_ with the following specifications:

- For points with target = 0, label should be 'Setosa' and color should be 'red'
- For points with target = 1, label should be 'Versicolor' and color should be 'blue'
- For points with target = 2, label should be 'Virginica' and color should be 'green'
- X axis should be labelled 'Petal Length'
- Y axis should be labelled 'Petal Width'
- Legend should be at top right corner


```python
colors = {0: 'red', 1: 'blue', 2: 'green'}
labels = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}

plt.figure(figsize=(10, 6))

for target in [0, 1, 2]:
    subset = df[df['target'] == target]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'], 
                label=labels[target], 
                color=colors[target], 
                alpha=0.6, edgecolors='w', s=100)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.legend(loc='lower right')
plt.show()
```
